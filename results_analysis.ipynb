```python
J=0.045*(400+300)**(2/3)
Allesina = ((4*df_1['Aij_var_ALL']**2)/(1+df_1['Aij_ALL']))**2
Allesina_inverse = ((1+ df_1['Aij_ALL']**2)/(4*df_1['Aij_var_AP']))**2
Stone= (1+df_1['Aij_ALL'])**2/df_1['Aij_var_ALL']
Weighted_Ri= (df_1['RiA']* df_1['Current_sp_A']+df_1['Current_sp_P']* df_1['RiP'])/(df_1['Current_sp_A']+df_1['Current_sp_P'])
Weighted_Intro= (df_1['IntrodA']* df_1['Current_sp_A']+df_1['Current_sp_P']* df_1['IntrodP'])/(df_1['Current_sp_A']+df_1['Current_sp_P'])

comp = (Weighted_Ri*(Weighted_Intro**0.5)*((1- df_1['Aij_ALL'])**2/df_1['Aij_var_ALL']))**0.75
 
trav_wave = (Weighted_Ri**(2/3)*(Weighted_Intro**2/3)*((1+ df_1['Aij_ALL']**2)/df_1['Aij_var_ALL'])**2)**0.5

threq = (Weighted_Intro)*((Weighted_Ri**(2/3))*((1+df_1['Aij_ALL'])**2/df_1['Aij_var_ALL']))**0.5

df_1['Allesina']=Allesina
df_1['Allesina_inverse']=Allesina_inverse
df_1['new']=Stone
df_1['comp']=comp
df_1['trav']=trav_wave
df_1['threq']= threq
df_1['Weighted_Intro']=Weighted_Intro
df_1['Weighted_Ri']=Weighted_Ri
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Create the figure with adjusted figsize
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(16, 16), sharex=True, sharey='none')


data = [
    (df_1['Current_sp_A']+df_1['Current_sp_P'], df_1['new'], 'a', r"$S$ "),
    
    (df_1['Current_sp_A']+df_1['Current_sp_P'], J*df_1['comp'], 'b', r"$S$")
    ,(df_1['Current_sp_A']+df_1['Current_sp_P'], df_1['trav'], 'c', r"$S$ "),
    
    (df_1['Current_sp_A']+df_1['Current_sp_P'], df_1['threq'], 'd', r"$S$"),
    
]

# Define the subplot labels
subplot_labels = ['a', 'b','c','d']

# Iterate over the subplots and data
for i, (ax, (x, y, title, equation)) in enumerate(zip(axs.flat, data)):
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    line = slope * np.array(x) + intercept

    # Plot scatter plot
    ax.scatter(x, y,s=14, color='black',label='Data')

    # Plot regression line
    #ax.plot(x, line, color='red')
    # Add subplot label at the top left corner using 'top left' option
    #ax.text(0, 1.1, subplot_labels[i], transform=ax.transAxes, fontsize=18, fontweight='bold', va='top', ha='left')
    # Add R-squared value as text
    r_squared = r_value**2
    #ax.annotate(f'$R^2$ = {r_squared:.2f}', xy=(0.6, 0.1), xycoords='axes fraction', fontsize=16)
    print(r_squared,p_value)

    # Add custom equation of the regression line
    #ax.text(0.2, 0.8, equation, transform=ax.transAxes, fontsize=20, color='black')

    if i%2==0:
        # Add labels and legend
        ax.set_ylabel('Predicted richness', fontsize=20, color='black')

    # Add subplot label
    #ax.text(0.05, 0.95, subplot_labels[i], transform=ax.transAxes, fontsize=18, fontweight='bold', va='top')

    # Increase the font size of the tick labels
    ax.tick_params(axis='both', labelsize=18)

    # Bolden the tick labels
    ax.xaxis.set_tick_params(which='both', width=1)
    ax.yaxis.set_tick_params(which='both', width=1)

    # Add x-axis label to bottom row subplots
    ax.set_xlabel('Observed richness', fontsize=20, color='black')
    ax.set_title(title, fontsize=18, fontweight='bold', loc="left")

# Adjust spacing between subplots
plt.tight_layout()

# Save the plot as a PNG image
plt.savefig('rich_com.png', dpi=900)

# Show the plot
plt.show()
```
